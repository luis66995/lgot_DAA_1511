```python
#funcion  para recibir paramentros  en los nodos
class EnlazeNodo:
    def __init__(self, value, arriba = None, centro = None, abajo = None):
        self.data = value
        self.arriba = arriba
        self.centro = centro
        self.abajo = abajo

#objeto que representara la cabeza

Objhead = EnlazeNodo(20, EnlazeNodo(23, ' ', EnlazeNodo(57), ' '),
          EnlazeNodo(19, ' ', ' ', EnlazeNodo(67, ' ', EnlazeNodo(99), ' ')))

print(Objhead.centro.abajo.centro.data)
print(Objhead.arriba.centro.data)
```
